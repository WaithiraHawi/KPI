```python
import pandas as pd

file_path = "/content/CallCenter.xlsx"
excel_file = pd.ExcelFile(file_path)
# Viewing available sheets
sheet_names = excel_file.sheet_names
# Loading main sheet
df = pd.read_excel(file_path, sheet_name='Call Center WALLBOARD')
# Standardizing column names
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
# Converting to numeric where applicable
numeric_columns = [
    'No_of_Agents', 'Call_Presented', 'Call_Handled', 'Call_abounded',
    'Longest_Wait_Time_(sec)', 'Avg_Vote_to_Agents',
    'Average_TalkTime_(sec)', 'Max_TalkTime_(sec)', 'Average_HoldTime_(sec)'
]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
# Calculating KPIs
df['Call_Answer_Rate'] = df['Call_Handled'] / df['Call_Presented']
df['Call_Abandonment_Rate'] = df['Call_abounded'] / df['Call_Presented']
```

```python
# Creating date column and sort
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']])
df = df.sort_values('Date')
df[[
    'Date', 'Call_Presented', 'Call_Handled', 'Call_abounded',
    'Call_Answer_Rate', 'Call_Abandonment_Rate',
    'Average_TalkTime_(sec)', 'Longest_Wait_Time_(sec)',
    'Average_HoldTime_(sec)'
]].head()
```

```python
import matplotlib.pyplot as plt
# Call Answer Rate Over Time
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Call_Answer_Rate'], marker='o', label='Call Answer Rate')
plt.title('Call Answer Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Answer Rate')
plt.ylim(0.9, 1.01)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Call Abandonment Rate Over Time
plt.figure(figsize=(7, 5))
plt.plot(df['Date'], df['Call_Abandonment_Rate'], marker='o', color='red', label='Call Abandonment Rate')
plt.title('Call Abandonment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Abandonment Rate')
plt.ylim(0, 0.05)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Call Abandonment Rate Over Time with dynamic y-limit buffer
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Call_Abandonment_Rate'], marker='o', color='red', label='Call Abandonment Rate')
plt.title('Call Abandonment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Abandonment Rate')

# Dynamically setting y-limits with a 10% buffer above the max value
y_max = df['Call_Abandonment_Rate'].max()
plt.ylim(0, y_max * 1.1 if y_max > 0 else 0.01)

plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Talk Time vs Hold Time
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Average_TalkTime_(sec)'], marker='o', label='Avg Talk Time (sec)')
plt.plot(df['Date'], df['Average_HoldTime_(sec)'], marker='s', label='Avg Hold Time (sec)')
plt.title('Talk Time vs Hold Time Over Time')
plt.xlabel('Date')
plt.ylabel('Time (sec)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#longest wait time
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Longest_Wait_Time_(sec)'], marker='x', color='purple', label='Longest Wait Time (sec)')
plt.title('Longest Wait Time Over Time')
plt.xlabel('Date')
plt.ylabel('Wait Time (sec)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Total Talk Time = Average Talk Time  Call Handled
df['Total_Talk_Time_(sec)'] = df['Average_TalkTime_(sec)'] * df['Call_Handled']

# Agent Available Time = No. of Agents  8 hours  3600 seconds/hour
df['Agent_Available_Time_(sec)'] = df['No_of_Agents'] * 8 * 3600

# Agent Utilization Rate = Total Talk Time / Agent Available Time
df['Agent_Utilization_Rate'] = df['Total_Talk_Time_(sec)'] / df['Agent_Available_Time_(sec)']

# Average Handling Time (AHT) = Average Talk Time + Average Hold Time
df['Average_Handling_Time_(sec)'] = df['Average_TalkTime_(sec)'] + df['Average_HoldTime_(sec)']

# Calls Per Agent = Calls Handled / No. of Agents
df['Calls_Per_Agent'] = df['Call_Handled'] / df['No_of_Agents']

# Previewing computed KPIs
df[['Date', 'Call_Handled', 'No_of_Agents',
    'Agent_Utilization_Rate', 'Average_Handling_Time_(sec)',
    'Calls_Per_Agent']].head()
```
